```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
df_tbi = pd.read_excel("H:\CTSN\Pour_revised_manuscripts\Data_TBI_after_process.xlsx")
df_tbi.head()
```

```python
X = df_tbi[['Age', 'Sex', 'Marshall (t0)', 'ERBI A (t1)', 'CRS-R (t1)', 'RLAS (t1)', 'ERBI A (t1)', 
                         'DRS (t1)', 'Entry Diagnosis (t0)']]
X
y = df_tbi[['GOS-E (t2)']]
X
```

```python
feature_names = X.columns
```

# Forward

```python
# Initialize the Random Forest model for feature selection
model = RandomForestClassifier(n_estimators=200, random_state=42,max_depth=10, min_samples_split=2)
```

```python
# Initialize forward selection (SequentialFeatureSelector) with Random Forest
sfs = SequentialFeatureSelector(model, n_features_to_select = 5, direction='forward')  # Adjust n_features_to_select
```

```python
# Perform forward selection on the entire dataset
sfs.fit_transform(X, y)

# Get the selected feature names
selected_features = feature_names[sfs.get_support()]

# Print the selected features
print("Selected features: ", selected_features)
```

### ['Age', 'ERBI A (t1)', 'CRS-R (t1)', 'RLAS (t1)', 'DRS (t1)']

# Backward 

```python

# Initialize backward selection (SequentialFeatureSelector) with Random Forest
sfs_bw = SequentialFeatureSelector(model, n_features_to_select = 5, direction='backward')

# Perform backward selection on the entire dataset
sfs_bw.fit_transform(X, y)

# Get the selected feature names
selected_features_bw = feature_names[sfs_bw.get_support()]

# Print the selected features
print("Selected features: ", selected_features_bw)
```

### ['Age', 'CRS-R (t1)', 'RLAS (t1)', 'DRS (t1)', 'Entry Diagnosis (t0)']
